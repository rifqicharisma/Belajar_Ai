```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('upn.jpg')
kernel_3_3 = np.ones((3,3),np.float32)/9
kernel_5_5 = np.ones((5,5),np.float32)/25

#dst = cv.filter2D(img,cv.CV_8U,kernel_5_5)
dst_3_3 = cv.filter2D(img,cv.CV_8U,kernel_3_3,(-1,-1), delta = 0, borderType = cv.BORDER_DEFAULT)
dst_5_5 = cv.filter2D(img,cv.CV_8U,kernel_5_5,(-1,-1), delta = 0, borderType = cv.BORDER_DEFAULT)

plt.subplot(1, 3, 1),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 2),plt.imshow(dst_3_3),plt.title('Averaging 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 3),plt.imshow(dst_5_5),plt.title('Averaging 5x5')
plt.xticks([]), plt.yticks([])
plt.show()

cv.imshow('Original', img)
cv.imshow('Averaging Kernel 3x3', dst_3_3)
cv.imshow('Averaging Kernel 5x5', dst_5_5)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
from skimage.util import random_noise
from matplotlib import pyplot as plt

img = cv.imread('monas.png')
noise_img = random_noise(img, mode='s&p',amount=0.1)

noise_img = np.array(255*noise_img, dtype = 'uint8')

blur_average = cv.blur(noise_img,(3,3))
blur_gaussian = cv.GaussianBlur(noise_img,(3,3),0)
blur_median = cv.medianBlur(noise_img,3)

plt.subplot(3, 3, 1),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(3, 3, 2),plt.imshow(noise_img),plt.title('Original with Noise')
plt.xticks([]), plt.yticks([])
plt.subplot(3, 3, 3),plt.imshow(blur_average),plt.title('Average Filter')
plt.xticks([]), plt.yticks([])
plt.subplot(3, 3, 4),plt.imshow(blur_gaussian),plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(3, 3, 5),plt.imshow(blur_median),plt.title('Median Filter')
plt.xticks([]), plt.yticks([])
plt.show()

cv.imshow('Original', img)
cv.imshow('Original with Noise', noise_img)
cv.imshow('Smoothing Average Filter', blur_average)
cv.imshow('Smoothing Gaussian', blur_gaussian)
cv.imshow('Smoothing Median Filter', blur_median)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('monas.png',0)

sobelx_f = cv.Sobel(img,cv.CV_64F,1,0,ksize=3)
sobely_f = cv.Sobel(img,cv.CV_64F,0,1,ksize=3)
sobelx_8u = cv.Sobel(img,cv.CV_8U,1,0,ksize=3)
sobely_8u = cv.Sobel(img,cv.CV_8U,0,1,ksize=3)

magnitude = cv.magnitude(sobelx_f,sobely_f)
abs_sobel64f = np.absolute(magnitude)
sobel_8u = np.uint8(abs_sobel64f)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(sobelx_8u,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobely_8u,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobel_8u,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()

cv.imshow('Original', img)
cv.imshow('Sobel X', sobelx_8u)
cv.imshow('Sobel Y', sobely_8u)
cv.imshow('Sobel Magn', sobel_8u)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import pandas as pd
```

# Data Preparation

```python
data = pd.read_csv('train.csv') #memanggil data dengan format .csv
data.head() #menampilkan data 10 baris teratas
```

```python
data.shape
```

```python
#missing value data
data.isnull().sum()
```

```python
data = data.drop(['Cabin','Name','Ticket'],axis=1) #axis=1 menandakan kolom
data = data.dropna()
data.shape
```

```python
#missing value data
data.isnull().sum()
```

```python
data.info()
```

```python
data = pd.get_dummies(data) #numerisasi tipe data di kolom
data.head()
```

```python
data.info()
```

```python
#pisahkan target dari data
y = data['Survived'] #target
x = data.drop(['Survived'],axis=1) #predictor/fitur
```

```python
x.shape, y.shape
```

```python
# pisahkan data train dan test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

```python
x_train.shape, y_train.shape
```

# Modelling

```python
from sklearn.neighbors import KNeighborsClassifier #memanggil algoritma KNN
```

```python
model = KNeighborsClassifier(n_neighbors=7) #define parameter
model.fit(x_train, y_train) #learning process
```

# Evaluasi Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report #memanggil mertik evaluasi
```

```python
x_test.shape, y_test.shape
```

```python
prediksi = model.predict(x_test) #prediksi kelas dari masing2 data
print("akurasi:",accuracy_score(y_test,prediksi)) #evaluasi prediksi vs groundtruth
```

```python
print("confusion matrix:\n",confusion_matrix(y_test,prediksi))
```

```python
print(classification_report(y_test,prediksi))
```

```python
# tunning hyperparameter
param = [3,5,7,9]
for p in param:
  model = KNeighborsClassifier(n_neighbors=p) #define parameter
  model.fit(x_train, y_train) #learning process
  prediksi = model.predict(x_test) #prediksi kelas dari masing2 data
  hasil = accuracy_score(y_test,prediksi) #evaluasi prediksi vs groundtruth
  print(hasil)
```

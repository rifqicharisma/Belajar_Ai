## Latihan 7

Hitunglah *Mean Absolute Error* (MAE) dari list nilai prediksi terhadap list nilai aslinya

$mae = (\frac{1}{n})\sum_{i=1}^{n}\left | y_{i} - x_{i} \right |$

dimana:
* n = jumlah data
* yi = nilai prediksi ke-i
* xi = nilai aktual ke-i

```python
y = [1.2, 5, 8.7, 10, 15]
x = [1, 4.5, 9.2, 9, 14]
#Jawaban
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, x)
```

## Latihan 10

Buatlah kode untuk menghitung luas gabungan dari persegi panjang dan segitiga

No | Lebar | Panjang | Alas | Tinggi
--- | --- | --- | --- | ---
1 | 2 | 5 | 3 | 6
2 | 3 | 5 | 4 | 8
3 | 3 | 6 | 2 | 5
4 | 5 | 8 | 2 | 4
5 | 4 | 10 | 3 | 5

```python
#Jawaban
lebar = [2,3,3,5,4]
panjang = [5,5,6,8,10]
alas = [3,4,2,2,3]
tinggi = [6,8,5,4,5]

for i in range(len(lebar)):
    luas1 = lebar[i] * panjang[i]
    print("Luas persegi -",i+1, " = ", luas1)
    luas2 = (alas[i] * tinggi[i]) / 2
    print("Luas segitiga -",i+1, " = ", luas2)

    luasGab = luas1 + luas2
    print("Luas gabungan -",i+1, " = ", luasGab)
    print("\n")
```

```
#DATA CLEANING: hapus beberapa missing value ?, 
df['price'] = df['price'].replace(to_replace ="?",value =0) #replace ? jadi 0
df['price'] = df['price'].astype("int64") #jadikan object --> int64
```

```
#simple normalization
#SIMPLE FEAtURE SCALING: 0 - 1
df['length'] = df['length'] / df['length'].max()
```

```
#Min Max normalization
df['wheel-base'] = (df['wheel-base'] - df['wheel-base'].min() / df['wheel-base'].max() - df['wheel-base'].min())
```

```
#z score
df['width'] = (df['width'] - df['width'].mean()) / df['width'].std()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# memberikan nama tabel header
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv('iris.csv', names=names)
dataset.head()
```

```python
# memisahkan atribut dengan class
X = dataset.iloc[:, :-1].values # atribut
y = dataset.iloc[:, 4].values # class
print("Attribute : ")
print(X)
print("\nClass : ")
print(y)
```

```python
from sklearn.model_selection import train_test_split
# memisahkan data training dengan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0) #

# menghitung jumlah data training dan testing
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # membuat objek scaler
scaler.fit(X_train) # memfit data training ke objek scaler
X_train = scaler.transform(X_train) # memtransform data training dengan memakai objek scaler
X_test = scaler.transform(X_test) # memtransform data testing dengan memakai objek scaler

print(X_train)
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=4) # k = 4
classifier.fit(X_train, y_train) # training KNN
```

```python
# melakukan prediksi dengan menggunakan X_test
y_pred = classifier.predict(X_test)

print("Class data yang benar : ")
print(y_test)

print("\nData Klasifikasi Model : ")
print(y_pred)
```

```python
# skor akurasi
classifier.score(X_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("\nConfusion Matrix : ")
print(confusion_matrix(y_test, y_pred)) # menghitung jumlah data yang benar dan salah
print("\nClassification Report : ")
print(classification_report(y_test, y_pred)) # menghitung akurasi
```

```python
error = []

# calculate error for each K values between 1 and 40
for i in range(1, 40): # menghitung error untuk setiap nilai K
    knn = KNeighborsClassifier(n_neighbors=i) # membuat model dengan nilai K
    knn.fit(X_train, y_train) # meng-fit model dengan data training
    pred_i = knn.predict(X_test) # melakukan prediksi dengan data testing
    error.append(np.mean(pred_i != y_test)) # menghitung error

plt.figure(figsize=(12, 6)) # ukuran plot
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
            markerfacecolor='blue', markersize=10) # plot grafik
plt.title('Error Rate K Value') # judul plot
plt.xlabel('K Value') # label x
plt.ylabel('Mean Error') # label y
plt.show() # menampilkan plot
print("\nMean Error : ")
print(error) # menampilkan persentase error
```
